```python
import copy, math
import matplotlib.pyplot as plt
import numpy as np

```

```python

# 数据集
x_train = np.array([[0.5, 1.5], [1, 1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])

# 分离两个类别的数据
class_0 = x_train[y_train == 0]
class_1 = x_train[y_train == 1]

# 绘制散点图
plt.figure(figsize=(8, 6))
plt.scatter(class_0[:, 0], class_0[:, 1], color='red', label='Class 0')
plt.scatter(class_1[:, 0], class_1[:, 1], color='blue', label='Class 1')

# 设置图表标题和图例
plt.title('Classification Dataset Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# 显示图表
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# 创建Logistic回归模型
model = LogisticRegression()

# 训练模型
model.fit(x_train, y_train)

# 生成分类边界的点
x_min, x_max = x_train[:, 0].min() - 0.5, x_train[:, 0].max() + 0.5
y_min, y_max = x_train[:, 1].min() - 0.5, x_train[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

# 预测整个网格上的值
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 绘制结果
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(class_0[:, 0], class_0[:, 1], color='red', label='Class 0')
plt.scatter(class_1[:, 0], class_1[:, 1], color='blue', label='Class 1')

# 设置图表标题和图例
plt.title('Classification with Logistic Regression')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# 显示图表
plt.show()
```
